```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('sample.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
fft_image = np.fft.fft2(gray_image)
fft_centralised = np.fft.fftshift(fft_image)
plt.figure(figsize = (15, 10))
plt.subplot(1, 2, 1)
plt.title('Original Image', size = 12)
plt.axis('off')
plt.imshow(image, cmap = "gray")
plt.subplot(1, 2, 2)
plt.title('Centralised Spectrum with fourier transform', size = 12)
plt.axis('off')
plt.imshow(np.log(1 + np.abs(fft_centralised)), "gray")
```

```python
fft_decentral = np.fft.ifftshift(fft_centralised)
fft_inversed = np.fft.ifft2(fft_decentral)
plt.figure(figsize = (15, 10))
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title('Decentralised spectrum', size = 12)
plt.imshow(np.log(1 + np.abs(fft_decentral)), "gray")
plt.subplot(1, 2, 2)
plt.axis('off')
plt.title('Image after inverse fourier transform', size = 12)
plt.imshow(np.real(fft_inversed), "gray")
```
