```python
from PIL import Image
```

```python
im1 = Image.open('color.png')
height1, width1 = im1.size

image1 = im1.load()

gamma = 4
for i in range(height1):
    for j in range(width1):
        val = []
        for color in image1[i, j]:
            val.append(int(4 * 255 * (color / 255) ** gamma))
        image1[i, j] = tuple(val)

im1
```

```python
im2 = Image.open('sample.png')
height2, width2 = im2.size

image2 = im2.load()

gamma = 0.6
for i in range(height2):
    for j in range(width2):
        val = []
        for color in image2[i, j]:
            val.append(int(255 * (color / 255) ** gamma))
        image2[i, j] = tuple(val)
im2
```
